```python
import json

# Load the scraped data
with open("../data/kink_shoots.json", "r") as f:
    data = json.load(f)

print(f"Total shoots scraped: {len(data['shoots'])}")
print(f"Total IDs processed: {len(data['scraped_ids']) + len(data['not_found_ids'])}")
```

```python
from collections import Counter

# Find all shoots where Lorelei Lee is a performer
lorelei_shoots = [
    shoot for shoot in data["shoots"]
    if "Lorelei Lee" in shoot.get("performers", [])
]

print(f"Shoots featuring Lorelei Lee: {len(lorelei_shoots)}\n")

collaborators = Counter()
channels = Counter()
for shoot in lorelei_shoots:
    for performer in shoot['performers']:
        if performer != "Lorelei Lee":
            collaborators[performer] += 1
    channels[shoot['channel']] += 1

print("\nCollaborators:")
for collaborator, count in collaborators.most_common():
    print(f"{collaborator}: {count}")
print("\nChannels:")
for channel, count in channels.most_common():
    print(f"{channel}: {count}")
```
